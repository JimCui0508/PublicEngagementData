```python
import matplotlib.pyplot as plt

# Data
categories = ['Treadmill', 'Elliptical', 'Weightlifting', 'Yoga mat', 'Resistance bands']
values = [19, 8, 39, 17, 24]

# Plotting
fig, ax = plt.subplots()
ax.bar(categories, values)

# Customizing the plot
ax.set_title('Health Equipment in Fitness Routine')
ax.set_xlabel('Equipment')
ax.set_ylabel('Number of respondents')
ax.grid(axis='y')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Daily', 'Several times a week', 'Once a week', 'Rarely', 'Never']
sizes = [8, 35, 8, 2, 0]
explode = (0.1, 0, 0, 0, 0)  # explode the first slice

# Plot
fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax.axis('equal')

# Title
ax.set_title("How often do you incorporate health equipment into your fitness routine?")

# Show
plt.show()
```
